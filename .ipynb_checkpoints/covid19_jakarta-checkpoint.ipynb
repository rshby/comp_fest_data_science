## Import Library

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 32})
import seaborn as sns
sns.set()

from warnings import filterwarnings
filterwarnings("ignore")

import scipy.stats as stat
```

## Import Datasets

```python
df = pd.read_excel("dataset/Daily Update Data Agregat Covid-19 Jakarta.xlsx", date_parser="Tanggal")
df.columns = df.columns.str.replace(" ", "_").str.lower()
df.drop(columns=["belum_diketahui_(masih_verifikasi)", "menunggu_hasil", "tenaga_kesehatan_terinfeksi"], axis="columns", inplace=True)
df["jam"] = df["jam"].fillna(df["jam"].mode()[0])
df["month"] = pd.DatetimeIndex(df['tanggal']).month
df["year"] = pd.DatetimeIndex(df['tanggal']).year
```

3. Dari dataset yang disediakan, temukan nilai-nilai outlier yang ada (menggunakan variabel yang kalian tentukan).

```python
df.columns
```

```python
columns = ["total_pasien", "sembuh", "meninggal", "self_isolation", "masih_perawatan", "positif_harian", "positif_aktif", "sembuh_harian"]
fig, axes = plt.subplots(ncols=2, nrows=4, figsize=(14, 12),constrained_layout=True)
fig.suptitle("Grafik Data Numerik")
for kolom, ax in zip(columns, axes.flatten()):
    sns.distplot(df[kolom], ax=ax, bins=30, color="g")
plt.show()
```

```python
columns = ["total_pasien", "sembuh", "meninggal", "self_isolation", "masih_perawatan", "positif_harian", "positif_aktif", "sembuh_harian"]
fig, axes = plt.subplots(ncols=2, nrows=4, figsize=(10, 8),constrained_layout=True)
fig.suptitle("Grafik Data Numerik")
for kolom, ax in zip(columns, axes.flatten()):
    sns.boxplot(df[kolom], ax=ax)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(7, 6), constrained_layout=True)
_ = ax.set_title("Dist Plot Sebelum Menghapus Outlier")
_ = sns.distplot(df["self_isolation"], ax=ax, color="r")

plt.show()
```

```python
q1 = df["self_isolation"].quantile(0.25)
q3 = df["self_isolation"].quantile(0.75)
IQR = q3 - q1
batas_bawah = q1 - (1.5 * IQR)
batas_atas = q3 + (1.5 * IQR)
print(f"nilai Q1 = {q1}\nnilai Q3 = {q3}\nnilai IQR = {IQR}\nBatas Bawah = {batas_bawah}\nBatas Atas = {batas_atas}")
```

```python
df[(df["self_isolation"]>=batas_bawah) & (df["self_isolation"]<=batas_atas)]["self_isolation"].agg(["max", "min"])
```

```python
fig, ax = plt.subplots(figsize=(7, 6), constrained_layout=True)
_ = ax.set_title("Dist Plot Setelah Menghapus Outlier")
_ = sns.distplot(df[(df["self_isolation"]>=batas_bawah) & (df["self_isolation"]<=batas_atas)]["self_isolation"], ax=ax, color="r")

plt.show()
```

4. Dari dataset yang disediakan, usulkan dua buah variabel dan berikan analisis korelasi antara kedua variabel tersebut. Jelaskan apa kesimpulan
yang dapat diambil berdasarkan analisis kalian.

```python
fig, ax = plt.subplots(figsize=(10, 10))
_ = sns.heatmap(df.loc[:,"total_pasien":"sembuh_harian"].corr(), annot=True, cmap="magma", linewidths=2, ax=ax)
```

```python
#fig, ax = plt.subplots(ncols=1, figsize=(10, 5), constrained_layout=True)
sns.jointplot(x="positif_aktif", y="positif_harian", data=df, stat_func=stat.pearsonr, kind="reg")
sns.scatterplot(x="positif_aktif", y="positif_harian", data=df, hue="month")
plt.legend()
```

```python
df.groupby(["year", "month"])[["positif_harian", "positif_aktif", "sembuh"]].mean().sort_values(by=["positif_harian", "positif_aktif"], ascending=False)
```

```python
df_2020 = df[df["year"]==2020]
```

```python
df_2020.groupby("month")[["positif_aktif", "positif_harian", "sembuh"]].mean()
```

```python
fig,ax = plt.subplots(ncols=2, constrained_layout=True, figsize=(12, 5))
_ = fig.suptitle("Rata-rata kasus per bulan")
_ = ax[0].plot(df_2020.groupby("month")[["positif_aktif"]].mean(), label="positif aktif", linewidth=2)
_ = ax[0].plot(df_2020.groupby("month")[["positif_harian"]].mean(), label="positif harian", color="r", linewidth=2)
_ = ax[0].plot(df_2020.groupby("month")[["sembuh"]].mean(), label="sembuh", color="g", linewidth=2)
_ = ax[0].set_title("2020")
_ = ax[0].set_xlabel("bulan")
_ = ax[0].set_ylabel("jumlah kasus")
_ = ax[0].legend()
_ = ax[0].set_ylim([0,96000])


_ = ax[1].plot(df[df["year"]==2021].groupby("month")[["positif_aktif"]].mean(), label="positif aktif", linewidth=2)
_ = ax[1].plot(df[df["year"]==2021].groupby("month")[["sembuh"]].mean(), label="sembuh", color="g", linewidth=2)
_ = ax[1].plot(df[df["year"]==2021].groupby("month")[["positif_harian"]].mean(), label="positif harian", color="r", linewidth=2)
_ = ax[1].set_title("2021")
_ = ax[1].set_xlabel("bulan")
_ = ax[1].set_ylabel("jumlah kasus")
_ = ax[1].set_ylim([0,96000])
_ = ax[1].legend()
```
